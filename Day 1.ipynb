```python
# Defines a "repeat" function that takes 2 arguments
def repeat(s, exclaim = True):
    # result = s + s + s
    result = s * 3 # this is faster since only do one time of calculation
    if exclaim:
        result = result + '!!!'
    return result
```

```python
def main():
    print(repeat('Yay', False))
    print(repeat('Woo Hoo '))
```

```python
main()
```

```python
def main(name):
    if name == 'Guido':
        print(repeat(name) + '!!!')
    else:
        print(repeat(name))
```

```python
## import panda
import pandas as pd
```

```python
## import the data
melbourne_file_path = '/home/kinux/llm_model/Kaggle ML/melb_data.csv'
melbourne_data = pd.read_csv(melbourne_file_path) # read in the data
melbourne_summary = melbourne_data.describe() # show the data detail as like the head of the data 
```

```python
mean_row = melbourne_summary.loc['count','Rooms']
print(mean_row)
```

```python
melbourne_data = melbourne_data.dropna(axis = 0)
```

```python
y = melbourne_data.Price
melbourne_features = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']
X = melbourne_data[melbourne_features]
```

```python
from sklearn.tree import DecisionTreeRegressor

melbourne_model = DecisionTreeRegressor(random_state=1)
## random_state is like the set.seed()??? Ask
melbourne_model.fit(X, y)
```

```python
## Model Validation
## Using Mean Absolute Error as metric
from sklearn.metrics import PredictionErrorDisplay, mean_absolute_error
predicted_home_prices = melbourne_model.predict(X)
mean_absolute_error(y, predicted_home_prices)
```

```python
## Splite the data into train and test
from sklearn.model_selection import train_test_split

train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=0)
melbourne_model = DecisionTreeRegressor(random_state = 1)
melbourne_model.fit(train_X, train_y)
## The model has been trained after this step
val_predictions = melbourne_model.predict(val_X)
print(mean_absolute_error(val_y, val_predictions))
```
